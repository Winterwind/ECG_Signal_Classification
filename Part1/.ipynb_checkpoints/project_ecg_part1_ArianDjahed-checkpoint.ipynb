### Step 3: Noise reduction/filtering using fourier transform (30 points)


````python
#apply fft to x and obtain the spectrum X
#compute the absolute value of X, name it X_abs
#obtain frequencies using fftfreq, name it freqs
X = fft.fft(x)
X_abs = np.abs(X)
freqs = fft.fftfreq(len(x), sampling_interval)
````

````python
#plot freqs vs X_abs
fig, ax = plt.subplots()
ax.plot(freqs, X_abs, '.', label='Spectrum')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Magnitude of Frequency Component')
ax.set_title('Magnitude Spectrum')
ax.set_xlim(-300, 300)
ax.set_ylim(0, 150000)
````

A normal resting heart rate for adults is between 60 and 100 Hz. \
Based on the above figure, we can see that there are components with high frequencies (> 100Hz or < -100Hz) \
We will set those components to zeros in X
```
 for n in range(0, len(X))
     if (freqs[n] < - 100) or (freqs[n] > 100):
         X[n]=0 
```
The above for loop does the job, but it is inefficient

````python
#use one line of code to replace the above for loop
#do not use ; in your code
#do not use any loop 
#do not use list comprehension
X[(freqs < -100) | (freqs > 100)] = 0
````

````python
#recompute the absolute value of X, name it X_abs
X_abs = np.abs(X)
````

````python
#plot freqs vs X_abs
fig, ax = plt.subplots()
ax.plot(freqs, X_abs, '.', label='Spectrum')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Magnitude of Frequency Component')
ax.set_title('Magnitude Spectrum')
ax.set_xlim(-300, 300)
ax.set_ylim(0, 150000)
````

````python
#apply the inverse fourier transform to X, and obtain x_clean
x_clean = fft.ifft(X)
````

````python
#we only need the real part
x_clean=x_clean.real
````

````python
#plot the signal x_clean
fig, ax = plt.subplots()
ax.plot(t, x_clean, label='ECG')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal Amplitude')
ax.set_title('ECG Signal x_clean')
ax.grid(True)
````

### Step 4: Peak Detection using cross-correlation (20 points)
Perform peak detection using two kernels: h1 and h2 
```
h1 = [-1, 1, 0] # kernel-1 for peak detection
h2 = [0, 1, -1] # kernel-2 for peak detection
```
you will get zero point for this step if you use find_peaks in sicpy 

````python
#you code for peak detection
h1 = [-1, 1, 0] # kernel-1 for peak detection
h2 = [0, 1, -1] # kernel-2 for peak detection

# Perform cross-correlation with the kernels
corr_h1 = correlate(x_clean, h1)
corr_h2 = correlate(x_clean, h2)

print(f'{corr_h1}\n{corr_h2}')

# Detect peaks by finding where the correlation is maximum
peaks_h1 = np.where((corr_h1[:-1] < 0) & (corr_h1[1:] > 0))[0] + 1
peaks_h2 = np.where((corr_h2[:-1] < 0) & (corr_h2[1:] > 0))[0] + 1

# Combine the peaks detected by both kernels
peaks = np.union1d(peaks_h1, peaks_h2)
print(peaks)
````

````python
#plot the detected peaks and x_clean
fig, ax = plt.subplots()
ax.plot(t, x_clean, label='ECG')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal Amplitude')
ax.set_title('ECG Signal x_clean')
ax.grid(True)
ax.plot(t[peaks], x_clean[peaks], 'o', label='Detected Peaks')
````

We need to find the R-peaks from the detected peaks \
Select the peaks with amplitude values > threshold \
Write code to find the threshold \
It is NOT acceptable to simply set threshold to a manually defined number (e.g., threshold=0.02)

````python
#estimate the threshold
threshold = 0.5 * np.max(x_clean[peaks])
r_peaks = peaks[x_clean[peaks] > threshold]
threshold
````

````python
#plot the R-peaks and x_clean
fig, ax = plt.subplots()
ax.plot(t, x_clean, label='ECG')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal Amplitude')
ax.set_title('ECG Signal x_clean')
ax.grid(True)
ax.plot(t[r_peaks], x_clean[r_peaks], 'o', label='Detected Peaks')
````

### Step 5: Calculate heart rate (10 points)

````python
# RPeakIndexArray stores the indexes of the R-peaks in the ECG signal x_clean
# step-1: create a numpy array HeartRate that has the same shape as RPeakIndexArray
# step-2: in a range-based for loop: 
#         HeartRate[n]=60*fs/(RPeakIndexArray[n]-RPeakIndexArray[n-1])
#         It should be range(1, RPeakIndexArray.shape[0])
# step-3: set HeartRate[0] to be equal to HeartRate[1]
# step-4: plot the signal x_clean and the HeartRate curve
RPeakIndexArray = r_peaks
HeartRate = np.zeros_like(RPeakIndexArray)
for n in range(1, RPeakIndexArray.shape[0]):
    HeartRate[n] = 60 * sampling_freq / (RPeakIndexArray[n] - RPeakIndexArray[n - 1])
HeartRate[0] = HeartRate[1]

fix, ax = plt.subplots(2, 1, figsize=(10, 6))

ax[0].plot(t, x_clean, label='ECG')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Signal Amplitude')
ax[0].grid(True)
ax[0].plot(t[r_peaks], x_clean[r_peaks], 'o', label='Detected Peaks')

ax[1].plot(t[r_peaks], HeartRate, label='Heart Rate')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Heart Rate per minute')
ax[1].grid(True)
ax[1].set_ylim(0, 100)
````
