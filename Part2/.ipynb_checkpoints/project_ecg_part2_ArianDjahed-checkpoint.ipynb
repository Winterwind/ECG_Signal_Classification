```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from time import time
```

```python
fs=125  # sampling frequency
Ts=1/fs # sampling interval
N=187 # the number of timepoints
Duration=N*Ts # duration of a signal
t=np.linspace(0, Duration-Ts, N) # array of timepoints
```

### Step 1: Load data (6 points)

```python
#load 'ECG_dataX.csv' via pandas and name it X
#load 'ECG_dataY.csv' via pandas and name it Y
#each row of X is an ecg signal - only the amplitude values
#each element of Y is the class label: Y[n] is the class label of the signal X[n]
X = pd.read_csv('ECG_dataX.csv')
Y = pd.read_csv('ECG_dataY.csv')
```

```python
#convert the dataframe X to a numpy array
X = X.values
```

### Step 2: Split the data into a training set (X_train, Y_train, 80%) and a test set (X_test, Y_test, 20%)  - 1 point
set random_state=0 in train_test_split

```python
#your code to split the data using train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

### Step 3A: Apply KNeighborsClassifier (6 points)

```python
#create a list of possible values of the parameter n_neighbors, name it n_list
n_list = np.arange(1, 97, step=5)
```

```python
n_list
```

```python
#define param_grid that only has n_neighbors
param_grid = {'n_neighbors': n_list}
```

```python
param_grid
```

```python
#use GridSearchCV to perform five fold cross-validation to find the optimal parameter - n_neighbors
#create a GridSearchCV object for KNeighborsClassifier in this cell
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn_grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, return_train_score=True)
knn_grid_search
```

```python
# run/fit GridSearchCV and measure the time cost
start_time = time()
knn_grid_search.fit(X_train, Y_train)
knn_time_cost = time()-start_time
print(f'KNeighborsClassifier: GridSearchCV time cost is {knn_time_cost} seconds')
```

```python
#show GridSearchCV results
knn_grid_search.cv_results_
```

```python
#show the best parameter
knn_grid_search.best_params_
```

```python
#obtain the best model
knn_grid_search.best_estimator_
```

```python
#measure the classification accuracy on the training set, using the best model
knn_train_acc = knn_grid_search.best_estimator_.score(X_train, Y_train)
print(f'KNeighborsClassifier: Training set accuracy is {knn_train_acc}')
```

evaluate the best model on the test set (the "final exam")

```python
#measure the classification accuracy on the test set, using the best model
knn_test_acc = knn_grid_search.best_estimator_.score(X_test, Y_test)
print(f'KNeighborsClassifier: Test set accuracy is {knn_test_acc}')
```

### Step 3B: Apply DecisionTreeClassifier (6 points)

```python
#create a list of possible values of the parameter max_depth, name it max_depth_list
max_depth_list = np.arange(1, 92, step=10)
```

```python
max_depth_list
```

```python
#define param_grid that only has max_depth and random_state
param_grid = {'max_depth': max_depth_list, 'random_state': [0]}
```

```python
param_grid
```

```python
#use GridSearchCV to perform five fold cross-validation to find the optimal parameter - max_depth
#create a GridSearchCV object for DecisionTreeClassifier in this cell
from sklearn.tree import DecisionTreeClassifier
dtc_grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, return_train_score=True)
dtc_grid_search
```

```python
#run/fit GridSearchCV and measure the time cost 
start_time = time()
dtc_grid_search.fit(X_train, Y_train)
dtc_time_cost = time()-start_time
print(f'DecisionTreeClassifier: GridSearchCV time cost is {dtc_time_cost} seconds')
```

```python
#show GridSearchCV results
dtc_grid_search.cv_results_
```

```python
#show the best parameter
dtc_grid_search.best_params_
```

```python
#obtain the best model
dtc_grid_search.best_estimator_
```

```python
#measure the classification accuracy on the training set, using the best model
dtc_train_acc = dtc_grid_search.best_estimator_.score(X_train, Y_train)
print(f'DecisionTreeClassifier: training accuracy is {dtc_train_acc}')
```

```python
#measure the classification accuracy on the test set, using the best model
dtc_test_acc = dtc_grid_search.best_estimator_.score(X_test, Y_test)
print(f'DecisionTreeClassifier: test accuracy is {dtc_test_acc}')
```

### Step 3C: Apply RandomForest (6 points)

```python
#create a list of possible values of the parameter max_depth, name it max_depth_list
max_depth_list = np.arange(1, 92, step=10)
```

```python
max_depth_list
```

```python
#define param_grid that only has max_depth and random_state
param_grid = {'max_depth': max_depth_list, 'random_state': [0]}
```

```python
param_grid
```

```python
#use GridSearchCV to perform five fold cross-validation to find the optimal parameter - max_depth
#create a GridSearchCV object for RandomForest in this cell
from sklearn.ensemble import RandomForestClassifier
rfc_grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, return_train_score=True)
rfc_grid_search
```

```python
#run/fit GridSearchCV and measure the time cost 
start_time = time()
rfc_grid_search.fit(X_train, Y_train)
rfc_time_cost = time()-start_time
print(f'RandomForestClassifier: GridSearchCV time cost is {rfc_time_cost} seconds')
```

```python
#show GridSearchCV results
rfc_grid_search.cv_results_
```

```python
#show the best parameter
rfc_grid_search.best_params_
```

```python
#obtain the best model
rfc_grid_search.best_estimator_
```

```python
#measure the classification accuracy on the training set, using the best model
rfc_train_acc = rfc_grid_search.best_estimator_.score(X_train, Y_train)
print(f'RandomForestClassifier: Training set accuracy is {rfc_train_acc}')
```

```python
#measure the classification accuracy on the test set, using the best model
rfc_test_acc = rfc_grid_search.best_estimator_.score(X_test, Y_test)
print(f'RandomForestClassifier: Test set accuracy is {rfc_test_acc}')
```
